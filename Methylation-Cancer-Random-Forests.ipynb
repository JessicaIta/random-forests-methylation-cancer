## Análise com Random Forests no banco de dados de Methylation

```python
#Importando as bibliotecas
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_predict
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
data = pd.read_csv('methylation-cancer.csv')
data
```

```python
len(data)
```

```python
#Separando preditores e a classe
X = data.values[:, 0:4]
Y = data.values[:,5]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3)
```

```python
#Aplicando o algoritmo com critério gini
clf_gini = RandomForestClassifier(n_estimators=100, random_state=0, criterion = "gini", max_depth=5)
clf_gini.fit(X_train, y_train)
```

```python
#Aplicando o algoritmo com critério entropia
clf_entropia = RandomForestClassifier(n_estimators=100, random_state=0, criterion = "entropy", max_depth=5)
clf_entropia.fit(X_train, y_train)
```

```python
#Predição com critério gini_index
y_pred = clf_gini.predict(X_test)
y_pred
```

```python
#Predição com critério ganho de informação
y_pred_en = clf_entropia.predict(X_test)
y_pred_en
```

```python
#Acurácia com o critério gini_index
print ("Acurácia foi ", accuracy_score(y_test,y_pred)*100)
```

```python
print ("Acurácia foi ", accuracy_score(y_test,y_pred_en)*100)
```
